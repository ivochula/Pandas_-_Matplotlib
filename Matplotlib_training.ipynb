```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```


# Plot


```python
x = np.arange(0,5)
y = np.arange(0,5)

plt.plot(x, y, 'yh--', label = "xxx")    

x2 = np.arange(0,3.5,0.3)
plt.plot(x2[:4], x2[:4]**2, 'r', label = 'X^2')
plt.plot(x2[3:], x2[3:]**2, 'b')    
plt.title("First Graph", fontdict= {'fontname': "Miriam",'fontsize': 20,'color': "Red",})
plt.xlabel("Eixo X", fontdict= {'fontname': "Miriam",'fontsize': 20,'color': "Red",})
plt.ylabel("Eixo Y", fontdict= {'fontname': "Miriam",'fontsize': 20,'color': "Red",})

plt.xticks([0, 1, 2, 3, 6], ['January', 'February', 'March', 'April', 'May'], rotation=45)
plt.yticks([0, 1, 2, 3, 6], ['January', 'February', 'March', 'April', 'May'], rotation=45)

plt.legend()

plt.savefig("grafico_teste.png", dpi = 300)

plt.show()
```

# Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values, color = 'g', label = "A,B,C", edgecolor = "b")

bars[0].set_hatch('/')
bars[1].set_hatch('+')
bars[2].set_hatch('O')

plt.legend()

plt.show()
```

# Pie Chart

```python
labels = 'Carne', 'Peixe', 'Frutas', 'Legumes', "Massas", "Arroz", "Água", "Nada"
sizes = [15, 10, 30, 45, 22, 5, 17, 23]
sizes_array = (np.array(sizes)) < 20
explode_values = [0.2 if x == True else 0 for x in sizes_array]
colors = ['red', 'blue', 'orange', 'green', 'yellow', 'grey', 'purple','pink']

plt.pie(sizes, labels = labels, explode = explode_values, colors = colors, autopct='%1.2f%%', 
        shadow = 1, radius = 3, rotatelabels = 0)

# best, upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center
plt.legend( loc = "best") 

plt.show()
```
